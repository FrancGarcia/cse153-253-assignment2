```python
from miditoolkit import MidiFile
import os
import pretty_midi
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tqdm import tqdm
import seaborn as sns
```

```python
# I'm only using the training data for data analysis just to showcase
# the rough distribution or else it's going to take A LOT of time
train_dir = "C:/Users/Francisco/Spring 2025/CSE 253R/Assignment 2/assignment2/cse153-253-assignment2\Data/train"

all_pitches = []
all_durations = []
all_velocities = []

midi_files = [f for f in os.listdir(train_dir) if f.endswith('.mid')]
for file in tqdm(midi_files[:4000]):
    try:
        midi = pretty_midi.PrettyMIDI(os.path.join(train_dir, file))
        for inst in midi.instruments:
            for note in inst.notes:
                all_pitches.append(note.pitch)
                all_durations.append(note.end - note.start)
                all_velocities.append(note.velocity)
    except Exception as e:
        print(f"Error with {file}: {e}")

# Use this dataframe for any data analysis for hidden markov chain
df = pd.DataFrame({
    'Pitch': all_pitches,
    'Duration': all_durations,
    'Velocity': all_velocities
})
```

```python
pitch_classes = df['Pitch'] % 12
pitch_class_counts = pitch_classes.value_counts().sort_index()

labels = ['C', 'C#', 'D', 'D#', 'E', 'F', 'F#', 'G', 'G#', 'A', 'A#', 'B']
plt.figure(figsize=(6, 6))
plt.pie(pitch_class_counts, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Pitch Class Distribution')
plt.show()
```

```python
duration_bins = pd.cut(df['Duration'], bins=[0, 0.1, 0.25, 0.5, 1.0, 2.0, np.inf],
                       labels=['<0.1s', '0.1–0.25s', '0.25–0.5s', '0.5–1.0s', '1.0–2.0s', '>2.0s'])
duration_counts = duration_bins.value_counts().sort_index()

fig, ax = plt.subplots(figsize=(8, 8))
wedges, texts, autotexts = ax.pie(
    duration_counts,
    startangle=90,
    autopct=lambda p: f'{p:.1f}%' if p > 2 else '',  
    wedgeprops=dict(width=0.4)
)

ax.legend(wedges, duration_counts.index,
          title="Duration",
          loc="center left",
          bbox_to_anchor=(1, 0.5))

plt.title('Note Duration Distribution')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
sns.histplot(df['Velocity'], bins=32, kde=True, color='dodgerblue')
plt.title('Velocity Distribution')
plt.xlabel('Velocity (1–127)')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()
plt.show()
```
