```python
import matplotlib.pyplot as plt
import numpy as np
```

## Log-likelihood per Instrument

```python
INSTRUMENTS = ['p1', 'p2', 'tr', 'no']

# results derived from other notebooks
hmm_llh = [-5.165, -5.174, -2.169, -3.201]
markov_llh = [-3.715, -3.930, -3.196, -2.706]

# derived from paper
lstm_quartet_llh = [-3.87, -3.71, -2.45, -1.62]
```

```python
# Bar width and positions
x = np.arange(len(INSTRUMENTS))
bar_width = 0.25

# Plot
fig, ax = plt.subplots()
ax.bar(x - bar_width, hmm_llh, bar_width, label='HMM')
ax.bar(x, markov_llh, bar_width, label='Markov Chain')
ax.bar(x + bar_width, lstm_quartet_llh, bar_width, label='LSTM-Quartet')

# Labels and formatting
ax.set_xlabel('Instruments')
ax.set_ylabel('Log-Likelihood')
ax.set_title('Model Log-Likelihood per Instrument')
ax.set_xticks(x)
ax.set_xticklabels(INSTRUMENTS)
ax.legend()
ax.grid(True, linestyle='--', alpha=0.5)
ax.invert_yaxis()  # Flip y-axis

plt.tight_layout()
plt.show()
```

## KL Divergence of pitch

```python
hmm_kl = [0.039340996622072376, 0.06244394502672241, 0.17909712939797354, 0.00539675125873548]
chain_kl = [0.08232277652595449, 0.09822218660567458, 0.10465714078364963, 0.007510883348865858]
```

```python
x = np.arange(len(INSTRUMENTS))
width = 0.35

plt.figure(figsize=(8, 5))
plt.bar(x - width/2, hmm_kl, width, label='HMM', color='skyblue')
plt.bar(x + width/2, chain_kl, width, label='Markov Chain', color='salmon')

plt.xlabel('Instruments')
plt.ylabel('KL Divergence')
plt.title('KL Divergence per Instrument (Pitch Distribution)')
plt.xticks(x, INSTRUMENTS)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
